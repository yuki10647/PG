```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('TrainOfTitanic.csv')
test = pd.read_csv('TestOfTitanic.csv')
```

```python
#要素確認
train.head()
test.head()
```

```python
#サイズ確認
train_shape = train.shape
test_shape = test.shape

print(train_shape)
print(test_shape)
```

```python
#基本統計量確認
train.describe()
test.describe()
```

```python
#欠損の確認
def kesson_table(df):
    null_val = df.isnull().sum()
    percent = 100 * df.isnull().sum()/len(df)
    kesson_table = pd.concat([null_val, percent], axis=1)
    kesson_table_ren_columns = kesson_table.rename(columns = {0:'欠損数', 1:'%'})
    return kesson_table_ren_columns

kesson_table(train)
kesson_table(test)
```

```python
#事前処理
#１．欠損データを埋める(代理データ)
#２．カテゴリ変数(文字列)を数字に変換
```

```python
#１．欠損データを埋める(代理データ)
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna("S")
 
kesson_table(train)
```

```python
#２．カテゴリ変数(文字列)を数字に変換
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1
train["Embarked"][train["Embarked"] == "S" ] = 0
train["Embarked"][train["Embarked"] == "C" ] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2
 
train.head()
```

```python
#testも同様の処理
test["Age"] = test["Age"].fillna(test["Age"].median())
test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"] == "female"] = 1
test["Embarked"][test["Embarked"] == "S"] = 0
test["Embarked"][test["Embarked"] == "C"] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2
test.Fare[152] = test.Fare.median()     #testはFareに1つだけ欠損あった。fillnaでもok
 
test.head(10)
```

```python
#予測モデル：決定木
```

```python
from sklearn import tree
```

```python
#trainの目的変数と説明変数
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare"]].values
# 決定木の作成
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_one, target)
#testの説明変数
test_features = test[["Pclass", "Sex", "Age", "Fare"]].values
#予測
my_prediction = my_tree_one.predict(test_features)
```

```python
#サイズと中身の確認
my_prediction.shape
#print(my_prediction)
```

```python
#提出ファイル
PassengerId = np.array(test["PassengerId"]).astype(int)
my_solution = pd.DataFrame(my_prediction, PassengerId, columns = ["Survived"])
my_solution.to_csv("my_tree_one.csv", index_label = ["PassengerId"])
```

```python
#説明変数を増やしてみる
```

```python
# 追加項目も含める
features_two = train[["Pclass","Age","Sex","Fare", "SibSp", "Parch", "Embarked"]].values
#過学習を考慮してアーギュメントの設定
max_depth = 10
min_samples_split = 5
my_tree_two = tree.DecisionTreeClassifier(max_depth = max_depth, min_samples_split = min_samples_split, random_state = 1)
my_tree_two = my_tree_two.fit(features_two, target)
```

```python
#予測
test_features_2 = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
my_prediction_tree_two = my_tree_two.predict(test_features_2)
```

```python
#ファイル作成
PassengerId = np.array(test["PassengerId"]).astype(int)
my_solution_tree_two = pd.DataFrame(my_prediction_tree_two, PassengerId, columns = ["Survived"])
my_solution_tree_two.to_csv("my_tree_two.csv", index_label = ["PassengerId"])
```
