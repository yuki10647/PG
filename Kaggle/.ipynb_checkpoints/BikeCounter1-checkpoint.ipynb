## 学習モデル

```python
# Lasso回帰 可視化
alphas = 10**np.linspace(-5, 10, 500)

Lasso_cv = LassoCV(alphas=alphas, cv=10)
Lasso_cv.fit(train_0, train_y)
print(Lasso_cv.alpha_)


kf = KFold(n_splits=10, shuffle=True, random_state=1).split(train_0, train_y)
kf_ = list(kf)

cnt = 0
sum = 0
for train, test in kf_:
    cnt += 1
    from sklearn.linear_model import Lasso,LassoCV
    Lasso = Lasso(alpha=Lasso_cv.alpha_)
    Lasso.fit(train_0.loc[train], train_y.loc[train])
    Lasso_train_y_pred = Lasso.predict(train_0.loc[test])
    Lasso_train_y_pred[Lasso_train_y_pred < 0] = 0
    print(f"Fold:{cnt}")
    print('RMSLE:', np.sqrt(mean_squared_log_error(train_y.loc[test], Lasso_train_y_pred)))
    score = np.sqrt(mean_squared_log_error(train_y.loc[test], Lasso_train_y_pred))
    sum += score

mean_score = sum / 10
print(f"mean_score_RMSLE = {mean_score}")
```

```python
# Lasso回帰 分析
from sklearn.linear_model import Lasso,LassoCV

alphas = 10**np.linspace(-5, 10, 500)

Lasso_cv = LassoCV(alphas=alphas, cv=10)
Lasso_cv.fit(train_0, train_y)
print(Lasso_cv.alpha_)

Lasso = Lasso(alpha=Lasso_cv.alpha_)
Lasso.fit(train_0, train_y)
print(Lasso.score(train_0, train_y))
Lasso_train_y_pred = Lasso.predict(test_0)
print(Lasso_train_y_pred)


#確認
#pd.Series(Lasso_train_y_pred).min()
#調整
#sub0 = pd.concat([test_datetime, pd.Series(Lasso_train_y_pred)], axis=1)
#sub0.columns = ['datetime','count']
#sub0['count'][sub0['count'] < 0] = 0
#提出
#sub0.to_csv('bikelasso.csv', index = False)
```
